```python
# import mplot3d
from mpl_toolkits import mplot3d

# import numpy
import numpy as np

# import matplotlib
import matplotlib.pyplot as plt

# create an empty canvas
ax = plt.axes(projection='3d')
```

------------

#### Task 1: Create wireframe of hyperbolic paraboloid on the domain <-10,10>X<-10,10>

* hyperbolic paraboloid equation is: z = x*x - y*y

```python
# wireframe
# divide the inetrval <-10,10> into 500 parts
x = np.linspace(-10,10,500)
y = np.linspace(-10,10,500)

# a Cartesian product
X,Y = np.meshgrid(x,y)

Z = X*X - Y*Y

#  create the figure and axes
fig,ax = plt.subplots(figsize=(12.8,9.6))

# set a 3D projection
ax = plt.axes(projection='3d')

# a wireframe
#ax.plot_wireframe(X, Y, Z, color='r')
#plt.show()
ax.plot_surface(X, Y, Z, cmap="jet")
ax.contour(X, Y, Z, 50, cmap="jet", zdir="z")
```


-------------
#### Task 2: Create surface plot of eliptical paraboloid on the domain <-10,10>X<-10,10>

* eliptical paraboloid : z = x*x + y*y

```python
# surface plot
Z = X*X + Y*Y

#  create the figure and axes
fig,ax = plt.subplots(figsize=(12.8,9.6))

# set a 3D projection
ax = plt.axes(projection='3d')

# the surface
ax.plot_surface(X, Y, Z, cmap='jet')
plt.show()
```

#### Task 3: Create surface plot of eliptical paraboloid on the domain <-10,10>X<-10,10>

* Also create a contour projection on the plane z=0 (HINT: use the parameter 'zdir' of contour plot)

* eliptical paraboloid : z = x*x + y*y

```python
# contour plot
#  create the figure and axes
fig,ax = plt.subplots(figsize=(12.8,9.6))

# set a 3D projection
ax = plt.axes(projection='3d')

planez0 = np.zeros((500,500))
# the surface
ax.contour(X, Y, Z, 40, cmap='jet')

ax.contour(X, Y, planez0, 40, cmap='jet', zdir='z')
plt.show()
```
