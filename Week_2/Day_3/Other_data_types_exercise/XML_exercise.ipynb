# XML exercise

Using data from [**mondial database**](https://drive.google.com/file/d/14lFT4nWHgwN36ij4XZh6OUuup-K9qLgR/view?usp=sharing) find the answers to following questions:

1. 10 countries with the lowest infant mortality rates
2. 10 cities with the largest population
3. name and country of a) longest river, b) largest lake and c) airport at highest elevation

```python
import xml.etree.ElementTree as ET
```

```python
tree = ET.parse('mondial.xml')
root = tree.getroot()
root
```

```python
print(root.tag)
print(root.attrib)
print(len(root))
```

```python
countries = {
    "name": [],
    "infant_mortality":  [],
    "population": [],
    "cities": []
}

for country in root.findall("country"):
    print(country.find("name").text)
    countries["name"].append(country.find("name").text)
    if_mortality = country.find("infant_mortality")
    countries["infant_mortality"].append(None if if_mortality is None else float(if_mortality.text))
    countries["population"].append(int(country.findall("population")[-1].text))
    for city in country.findall("city"):
        print(city.find("name").text)
        cities = []
        try:
            cityDict = {
                "name": city.find("name").text,
                "population": int(city.findall("population")[-1].text)
            }
            cities.append(cityDict)
        except:
            print("no population")
            cities.append({
                "name": city.find("name").text,
                "population": None})
    countries["cities"].append(cities)
```

```python
import pandas as pd
df = pd.DataFrame(countries) 
```

```python
df.head()
```

```python
#10 countries with the lowest infant mortality rates
df.sort_values(by="infant_mortality").head(10)
```

```python
#10 cities with the largest population
citiesdf = pd.json_normalize(df["cities"])[0].apply(pd.Series)
citiesdf["population"] = pd.to_numeric(citiesdf["population"])
citiesdf.sort_values("population")
```
